```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load 
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas Data Frames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset
school_data_complete = pd.merge(student_data, school_data, how="left", 
                                left_on = "school", right_on = "name")
# Rename columns names 
school_data_complete.rename(columns={"name_x":"StudentName","name_y":"SchoolName"}, inplace=True)

school_student_data = school_data_complete
#School_student_data
```

## School Summary

* Create an overview table that summarizes key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)
  
* Create a dataframe to hold the above results

```python
#School Name
School_Name = school_student_data["SchoolName"].unique().tolist()
#print(School_Name)
#School Type
#School_Type = school_student_data["type"].unique().tolist()
School_Type = school_data["type"].tolist()
#print(School_Type)
#Total Students
School_Size = school_data["size"]

#Total School Budget
School_Budget = school_data["budget"]

#Per Student Budget
Per_Stu_Budget = school_data["budget"]/school_data["size"]
#print(Per_Stu_Budget)
#Average Math Score
Avg_math_score = round(school_student_data.groupby(["SchoolName"])["math_score"].mean(),2).tolist()
#print(Avg_math_score)
#Average Reading Score
Avg_read_score = round(school_student_data.groupby(["SchoolName"])["reading_score"].mean(),2).tolist()
#print(Avg_read_score)

#Total students per school 
Total_Stu_Per_School = school_student_data.groupby(["SchoolName"])["Student ID"].count()

Math_score = school_student_data[school_student_data["math_score"] >= 70]

Reading_score = school_student_data[school_student_data["reading_score"] >= 70]

#Total students per school or grouped by school who passed with score >= 70
School_Math_Pass = Math_score.groupby(["SchoolName"])["Student ID"].count()
School_Read_Pass = Reading_score.groupby(["SchoolName"])["Student ID"].count()

#% Passing Math
Per_Pass_Math = (School_Math_Pass/Total_Stu_Per_School*100)
Fomatted_Per_Pass_Math = Per_Pass_Math.map('{:.2f}%'.format).tolist()

#% Passing Reading
Per_Pass_Reading = (School_Read_Pass/Total_Stu_Per_School*100)
Formatted_Per_Pass_Reading = Per_Pass_Reading.map('{:.2f}%'.format).tolist()

#Overall Passing Rate (Average of the above two)
Overall_Per_Rate = (Per_Pass_Math+Per_Pass_Reading)/2
Formatted_Overall_Per_Rate = Overall_Per_Rate.map('{:.2f}%'.format).tolist()

School_Summary = pd.DataFrame({"School Name" : School_Name,
                               "School Type" : School_Type,
                                "Total Students" : School_Size,
                                "Total School Budget" : School_Budget,
                                "Per Student Budget" : Per_Stu_Budget,
                                "Average Math Score" : Avg_math_score,
                                "Average Read Score" : Avg_read_score,
                                 "% Passing Math" : Fomatted_Per_Pass_Math,
                                 "% Passing Reading" : Formatted_Per_Pass_Reading,
                                 "Overall Passing Rate" : Formatted_Overall_Per_Rate      
                                })

#Optional: give the displayed data cleaner formatting
School_Summary = School_Summary[["School Name","School Type","Total Students","Total School Budget",
                                "Per Student Budget","Average Math Score","Average Read Score",
                                "% Passing Math","% Passing Reading","Overall Passing Rate"  ]]
School_Summary.set_index("School Name")
```

## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
# Sample bins
spending_bins = [0, 585, 615, 645, 675]
group_names = ["<$585", "$585-615", "$615-645", "$645-675"]

#Create a new dataframe with per student budget by appending the existing school data with the same field
# and merging the same with the student data

school_data["Per Student Budget"] = Per_Stu_Budget
school_student_data_bin = pd.merge(student_data, school_data, how="left", left_on = "school", 
                                   right_on = "name")
school_student_data_bin.rename(columns={"name_x":"Student_Name", "name_y":"School_Name"},inplace=True)
school_student_data_bin["SchoolSpendingRange"] = pd.cut(school_student_data_bin["Per Student Budget"], spending_bins, labels=group_names)

#Average Math Score
Avg_math_score_bybin = school_student_data_bin.groupby(["SchoolSpendingRange"])["math_score"].mean().tolist()

#Average Reading Score
Avg_reading_score_bybin = school_student_data_bin.groupby(["SchoolSpendingRange"])["reading_score"].mean().tolist()

#% Passing Math
Math_scoring_GTEQ70 = school_student_data_bin[school_student_data_bin["math_score"] >= 70]
Math_scoring_GTEQ70_bybin = Math_scoring_GTEQ70.groupby(["SchoolSpendingRange"])["Student ID"].count()

Total_students_bybin = school_student_data_bin.groupby(["SchoolSpendingRange"])["Student ID"].count()

Percent_Pass_Math_bybin = ((Math_scoring_GTEQ70_bybin/Total_students_bybin)*100)
Formatted_Percent_Pass_Math_bybin = Percent_Pass_Math_bybin.map('{:.2f}%'.format).tolist()

#% Passing Read
# (Students scoring >= 70 in a bin / Total students that particular bin) * 100
Read_scoring_GTEQ70 = school_student_data_bin[school_student_data_bin["reading_score"] >= 70]
Read_scoring_GTEQ70_bybin = Read_scoring_GTEQ70.groupby(["SchoolSpendingRange"])["Student ID"].count()

Total_students_bybin = school_student_data_bin.groupby(["SchoolSpendingRange"])["Student ID"].count()

Percent_Pass_Read_bybin = ((Read_scoring_GTEQ70_bybin/Total_students_bybin)*100)
Formatted_Percent_Pass_Read_bybin = Percent_Pass_Read_bybin.map('{:.2f}%'.format).tolist()

#Overall Passing Rate
Overall_Pass_Rate_bybin = (Percent_Pass_Math_bybin + Percent_Pass_Read_bybin)/2

Formatted_Overall_Pass_Rate_bybin = Overall_Pass_Rate_bybin.map('{:.2f}%'.format).tolist()

Scores_by_school_spending = pd.DataFrame({"Spending Ranges (Per Student)":group_names,
                                          "Average Math Score":Avg_math_score_bybin,
                                          "Average Reading Score":Avg_reading_score_bybin,
                                          "% Passing Math":Formatted_Percent_Pass_Math_bybin,
                                          "% Passing Reading": Formatted_Percent_Pass_Read_bybin,
                                          "% Overall Passing Rate": Formatted_Overall_Pass_Rate_bybin
                                         })
Scores_by_school_spending = Scores_by_school_spending[["Spending Ranges (Per Student)",
                                                      "Average Math Score","Average Reading Score",
                                                      "% Passing Math","% Passing Reading",
                                                      "% Overall Passing Rate"]]
Scores_by_school_spending.set_index("Spending Ranges (Per Student)")
```
